```python
import os
import json
import pandas as pd
```

```python
import mysql.connector as sql

my_db = sql.connect(
    host="localhost",
    user="root",
    password="12345678",
     database="phonepay_pulse"
)

cursor = my_db.cursor()
mycursor = my_db.cursor(buffered=True)

# Create database
cursor.execute("CREATE DATABASE IF NOT EXISTS phonepay_pulse")

print("Database created successfully!")
```

```python

# -------------------- CREATE TABLE --------------------
mycursor.execute("""
CREATE TABLE IF NOT EXISTS aggregated_transaction (
    State VARCHAR(255),
    Year INT,
    Quarter INT,
    Transaction_type VARCHAR(255),
    Transaction_count BIGINT,
    Transaction_amount DOUBLE
)
""")

print("Table created ")

# -------------------- PATH --------------------
path = r"D:\Guvi\Mini-project _1_(Python+SQL)\phonepay_pulse\pulse\data\aggregated\transaction\country\india\state\\"

# -------------------- DATA EXTRACTION --------------------
data_list = []

for state in os.listdir(path):
    state_path = os.path.join(path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            file_path = os.path.join(year_path, file)

            with open(file_path, 'r') as f:
                data = json.load(f)

                # check if data exists
                if data.get('data') and data['data'].get('transactionData'):
                    for item in data['data']['transactionData']:
                        data_list.append((
                            state,
                            int(year),
                            int(file.strip(".json")),
                            item['name'],
                            item['paymentInstruments'][0]['count'],
                            item['paymentInstruments'][0]['amount']
                        ))

print("Data extracted ")

# -------------------- INSERT INTO MYSQL --------------------
query = """
INSERT INTO aggregated_transaction
(State, Year, Quarter, Transaction_type, Transaction_count, Transaction_amount)
VALUES (%s, %s, %s, %s, %s, %s)
"""

mycursor.executemany(query, data_list)
my_db.commit()

print("Data inserted successfully ")
```

# 1. Decoding Transaction Dynamics on PhonePe

#   Query:1.Total Transactions Over Years 

```python
query = '''SELECT Year,
       SUM(Transaction_count) AS Total_Transactions
FROM aggregated_transaction
GROUP BY Year
ORDER BY Year'''

cursor.execute(query)

result = cursor.fetchall()

# Convert to DataFrame
df = pd.DataFrame(result, columns=[
    "Year",
    "Total_Transactions"
])

print(df)
```

```python
import plotly.express as px
fig = px.line(
    df,
    x="Year",
    y="Total_Transactions",
    title="Total Transactions Over Years",
    markers=True,
    line_shape="spline",             
    color_discrete_sequence=["#2E86AB"],
    width=900,
    height=500
)
fig.show()
```

# Query:2.Total Transaction Amount Over Years 

```python
query = """
SELECT Year,
SUM(Transaction_amount) AS Total_Amount
FROM aggregated_transaction
GROUP BY Year
ORDER BY Year
"""

cursor.execute(query)
data = cursor.fetchall()

df = pd.DataFrame(data, columns=["Year","Total_Amount"])
print(df)
```

```python
import plotly.express as px
fig = px.line(
    df,
    x="Year",
    y="Total_Amount",
    title="Total Transaction Amount Over Years",
    markers=True,
    line_shape="spline",             
    color_discrete_sequence=["#2E86AB"],
    width=900,
    height=500
)
fig.update_traces(
    marker=dict(size=6),
    line=dict(width=2))
fig.show()
```

# Query:3.Top 10 States by Transaction Amount 

```python
query ='''SELECT State,
SUM(Transaction_amount) AS Total_Amount
FROM aggregated_transaction
GROUP BY State
ORDER BY Total_Amount DESC
LIMIT 10''' 

cursor.execute(query)
result = cursor.fetchall()

df = pd.DataFrame(result,columns=[
                  "State",
                  "Total_Amount"])
print(df)
```

```python
import plotly.express as px
colors = [
    "#636EFA",  # blue
    "#EF553B",  # red
    "#00CC96",  # green
    "#AB63FA",  # purple
    "#FFA15A",  # orange
    "#19D3F3",  # cyan
    "#FF6692",  # pink
    "#B6E880",  # light green
    "#FF97FF"   # magenta
]
fig = px.bar(
    df,
    x="State",
    y="Total_Amount",
    color= "State",
    title="Top 10 States by Transaction Amount",
    color_discrete_sequence= colors
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white",
    width=1000,
    height=500,
    bargap=0.2,
    xaxis_tickangle=-45
)

fig.show()
```

# Query:4.Transaction Type Distribution 

```python
query = """
SELECT Transaction_type,
SUM(Transaction_count) AS Total_Transactions
FROM aggregated_transaction
GROUP BY Transaction_type
ORDER BY Total_Transactions DESC
"""

cursor.execute(query)
data = cursor.fetchall()

df = pd.DataFrame(data, columns=["Transaction_type","Total_Transactions"])
print(df)
```

# Query:5.Top states transaction trend

```python
query = """
SELECT State, Year,
SUM(Transaction_amount) AS Total_Amount
FROM aggregated_transaction
GROUP BY State, Year
ORDER BY State, Year
"""

cursor.execute(query)
data = cursor.fetchall()

import pandas as pd

df = pd.DataFrame(data, columns=["State","Year","Total_Amount"])

print(df["State"].unique())
```

```python
top_states = df.groupby("State")["Total_Amount"].sum()

top_states = top_states.sort_values(ascending=False).head(5)

top_states_list = list(top_states.index)

print(top_states_list)
```

```python
df_top = df[df["State"].isin(top_states_list)]

print(df_top["State"].unique())
```

```python
fig = px.line(
    df_top,
    x="Year",
    y="Total_Amount",
    color="State",
    markers=True,
    title="Top 5 States Transaction Trend",
    width=900,
    height=500
)

fig.show()
```

```python
query = """
SELECT User_brand,
SUM(User_count) AS Total_Users
FROM aggregated_users
GROUP BY User_brand
ORDER BY Total_Users DESC
LIMIT 10
"""

cursor.execute(query)
data = cursor.fetchall()

import pandas as pd

df = pd.DataFrame(data, columns=["Brand","Total_Users"])

print(df)
```

# Query-2:Device Brands by User Percentage

```python
query = """
SELECT User_brand,
AVG(User_percentage) AS Avg_User_Percentage
FROM aggregated_users
GROUP BY User_brand
ORDER BY Avg_User_Percentage DESC
LIMIT 10
"""

cursor.execute(query)
data = cursor.fetchall()

import pandas as pd
df = pd.DataFrame(data, columns=["Brand","Avg_Percentage"])

print(df)
```

Find states where:

Registered users are high

Users installed PhonePe but do not use it frequently.

```python
query = """ 
SELECT 
    State,
    SUM(Registered_user) AS Total_Users,
    SUM(App_opens) AS Total_App_Opens,
    SUM(App_opens) / SUM(Registered_user) AS Engagement_Ratio
FROM map_users
GROUP BY State
ORDER BY Engagement_Ratio ASC
LIMIT 10""" 
cursor.execute(query)
data = cursor.fetchall()

df = pd.DataFrame(data, columns=[
    "State",
    "Total_Users",
    "Total_App_Opens",
    "Engagement_Ratio"
])

print(df.to_string(float_format='{:,.2f}'.format))
```
